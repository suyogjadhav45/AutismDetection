```python
#Creating the dependent variable class
factor = pd.factorize(df['UserGender'])
df.UserGender = factor[0]
definitions = factor[1]
print(df.UserGender.head())
print(definitions)
```

```python
#Creating the dependent variable class
factor = pd.factorize(df['MediaName'])
df.MediaName = factor[0]
definitions = factor[1]
print(df.MediaName.head())
print(definitions)
```

```python
#Creating the dependent variable class
factor = pd.factorize(df['Group'])
df.Group = factor[0]
definitions = factor[1]
print(df.Group.head())
print(definitions)
```

```python
df['Media ID'] = df['Media ID'].astype(float)
df['MediaName'] = df['MediaName'].astype(float)
df[' Media Duration (sec - U=UserControlled)'] = df[' Media Duration (sec - U=UserControlled)'].astype(float)
df[' AOI ID'] = df[' AOI ID'].astype(float)
df[' AOI Start'] = df[' AOI Start'].astype(float)
df[' AOI Duration (sec - U=UserControlled)'] = df[' AOI Duration (sec - U=UserControlled)'].astype(float)
```

```python
df[' User ID'] = df[' User ID'].astype(float)
df['UserGender'] = df['UserGender'].astype(float)
df[' User Age'] = df[' User Age'].astype(float)
df[' Fixations (#)'] = df[' Fixations (#)'].astype(float)
df[' Revisits (#)'] = df[' Revisits (#)'].astype(float)
df['Group'] = df['Group'].astype(float)
```

```python
df.dtypes
```

```python
# Create X (feature matrix)
X = df.drop("Group", axis=1)

#Create Y (labels)
y = df["Group"]
```

```python
np.random.seed(42)

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

clf.fit(X_train, y_train);
```

```python
clf.score(X_test, y_test)
```

```python
#make prediciton
y_pred = clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
import pickle

pickle.dump(clf, open("./autism_detection.pkl", "wb"))
```

```python
np.random.seed(42)

from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=4)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

clf.fit(X_train, y_train);
```

```python
clf.score(X_test, y_test)
```

```python
#make prediciton
y_pred = clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
